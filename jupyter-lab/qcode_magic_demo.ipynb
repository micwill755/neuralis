## Load the Extension

Now load the extension in your notebook:

```python
%load_ext qcode_magic
```

## Using the Line Magic

You can use `%qcode` followed by your question on the same line:

```python
%qcode How do I create a scatter plot in matplotlib?
```

## Using the Cell Magic

For longer questions, you can use `%%qcode` and write your question on subsequent lines:

```python
%%qcode
How do I create a pandas DataFrame from a dictionary and filter rows based on a condition?
Can you show me an example with sample data?
```

## Try It Yourself

Now you can ask your own questions using the `%qcode` magic command:

```python
%%qcode 
can you write me code that will create a sample scatter plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='Color Value')
plt.title('Scatter Plot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, alpha=0.3)
plt.show()
```
